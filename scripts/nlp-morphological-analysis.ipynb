# Morphological Analysis in Python using library polyglot


### Morphological parsing

- In natural language processing, is the process of determining the morphemes from which a given word is constructed. 
- It must be able to distinguish between orthographic rules and morphological rules. 
- For example, the word 'foxes' can be decomposed into 'fox' (the stem), and 'es' (a suffix indicating plurality).

# Find morphemes of words and tokenizing using Morphological Analysis

```python
from polyglot.text import Text, Word
```

```python
words = ["preprocessing", "processor", "invaluable", "thankful", "crossed"]
for w in words:
  w = Word(w, language="en")
  print("{:<20}{}".format(w, w.morphemes))
```

```python
''' Splitting string by morphological analysis'''
blob = "Wewillmeettoday."
text = Text(blob)
text.language = "en"
text.morphemes
```
